```python
def detectar_fechas(df, sample_frac=1.0, min_parse_ratio=0.8):
    """Detecta columnas de fecha intentando parsear; devuelve lista de nombres."""
    fecha_cols = []
    cols_obj = df.select_dtypes(include=["object", "string"]).columns.tolist()
    sample_df = df.sample(frac=min(sample_frac, 1.0), random_state=1337) if len(df) > 0 else df
    for col in cols_obj:
        parsed = pd.to_datetime(sample_df[col], errors="coerce", infer_datetime_format=True, utc=False)
        ratio = parsed.notna().mean() if len(sample_df) else 0
        if ratio >= min_parse_ratio:
            fecha_cols.append(col)
    return fecha_cols
```

```python
def resumen_categoricas(df, topn=10):
    cat_cols = df.select_dtypes(include=["object", "category", "string", "bool"]).columns
    rows = []
    detalles = {}
    for c in cat_cols:
        vc = df[c].astype("string").value_counts(dropna=False)
        top = vc.head(topn)
        filas_top = pd.DataFrame({
            "valor": top.index.astype(str),
            "conteo": top.values,
            "porcentaje": (top.values / len(df)) * 100 if len(df) else 0
        })
        detalles[c] = filas_top
        fila = {
            "columna": c,
            "n_unicos": df[c].nunique(dropna=True),
            "mas_frecuente": top.index[0] if len(top) else np.nan,
            "freq_mas_frecuente": int(top.values[0]) if len(top) else np.nan,
            "pct_mas_frecuente": (top.values[0]/len(df))*100 if len(df) else np.nan
        }
        rows.append(fila)
    base = pd.DataFrame(rows).sort_values("pct_mas_frecuente", ascending=False) if rows else pd.DataFrame(
        columns=["columna", "n_unicos", "mas_frecuente", "freq_mas_frecuente", "pct_mas_frecuente"]
    )
    return base, detalles
```

```python
def stats_numericas(df):
    num_cols = df.select_dtypes(include=[np.number]).columns
    if len(num_cols)==0:
        return pd.DataFrame(columns=["columna","conteo","media","desv_std","min","p25","p50","p75","max","n_miss","pct_miss","skew","kurtosis"])
    desc = df[num_cols].describe(percentiles=[0.25,0.5,0.75]).T.rename(
        columns={"count":"conteo","mean":"media","std":"desv_std","min":"min","25%":"p25","50%":"p50","75%":"p75","max":"max"}
    )
    desc["n_miss"] = df[num_cols].isna().sum()
    desc["pct_miss"] = (desc["n_miss"]/len(df))*100 if len(df) else 0
    desc["skew"] = df[num_cols].skew(numeric_only=True)
    desc["kurtosis"] = df[num_cols].kurt(numeric_only=True)
    desc.insert(0, "columna", desc.index)
    desc.reset_index(drop=True, inplace=True)
    return desc
```

```python
def missing_por_col(df):
    if len(df)==0:
        return pd.DataFrame(columns=["columna","missing","pct_missing"])
    miss = df.isna().sum().sort_values(ascending=False)
    out = pd.DataFrame({"columna": miss.index, "missing": miss.values})
    out["pct_missing"] = (out["missing"]/len(df))*100
    return out
```

```python
def plot_histograma(serie, titulo, ruta_salida):
    plt.figure()
    data = pd.to_numeric(serie.dropna(), errors="coerce").dropna()
    if len(data) == 0:
        plt.close()
        return False
    plt.hist(data, bins=30)
    plt.title(titulo)
    plt.xlabel("Valor")
    plt.ylabel("Frecuencia")
    plt.tight_layout()
    plt.savefig(ruta_salida)
    plt.close()
    return True
```

```python
def plot_barras(series_counts, titulo, ruta_salida, rot=45):
    plt.figure()
    series_counts.plot(kind="bar")
    plt.title(titulo)
    plt.xlabel("Categoría")
    plt.ylabel("Conteo")
    plt.xticks(rotation=rot, ha="right")
    plt.tight_layout()
    plt.savefig(ruta_salida)
    plt.close()
    return True
```

```python
def plot_serie_tiempo(fecha_series, titulo, ruta_salida, freq="D"):
    fechas = pd.to_datetime(fecha_series, errors="coerce")
    fechas = fechas.dropna()
    if len(fechas)==0:
        return False
    conteo = fechas.dt.to_period(freq).value_counts().sort_index()
    idx = conteo.index.astype(str)
    plt.figure()
    plt.plot(range(len(conteo)), conteo.values)
    plt.title(titulo)
    plt.xlabel("Tiempo")
    plt.ylabel("Conteo")
    # etiquetas discretas para que no se amontonen
    paso = max(1, len(idx)//12)
    xticks_pos = list(range(0, len(idx), paso))
    xticks_lbl = [idx[i] for i in xticks_pos]
    plt.xticks(xticks_pos, xticks_lbl, rotation=45, ha="right")
    plt.tight_layout()
    plt.savefig(ruta_salida)
    plt.close()
    return True
```
